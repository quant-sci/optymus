# Constraint optimization

```python
# import libraries
import jax.numpy as jnp
from optymus import Optimizer
```

```python
from optymus.benchmark import Booth

f = Booth()

def constraint1(x):
    return 10*x[1] + 4*x[0] - 10
```

```python
initial_guess = jnp.array([0.0, -7.0]) 
cons_list = [constraint1]

opt = Optimizer(f_obj=f, f_cons=cons_list, x0=initial_guess, method='steepest_descent')
```

```python
opt.report()
```
